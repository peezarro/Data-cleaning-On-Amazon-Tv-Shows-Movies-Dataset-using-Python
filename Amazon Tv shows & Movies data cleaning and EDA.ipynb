```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plot the distribution of movie and TV show types
sns.countplot(data=df, x='type')
plt.title('Count on Movies and Tv Shows')
plt.show
```

```python
sns.countplot(data=df, x = 'rating')
plt.title('Count On Rating')
plt.show()
```

```python
# Plot the distribution of ratings
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='rating', order = df['rating'].value_counts().index)
plt.title('Distribution of Rating')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the distribution of release years
plt.figure(figsize=(12, 6))
sns.histplot(df['release_year'], bins=30, kde=False)
plt.title('Distribution of Release Years')
plt.show()
```

```python
# Relationship between release year and type
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='release_year', hue='type', order=df['release_year'].value_counts().index)
plt.title('Release Year vs. Type')
plt.xticks(rotation=90)
plt.show()
```

```python
# Relationship between ratings and type
plt.figure(figsize=(12, 6))
sns.countplot(data=df, x='rating', hue='type', order=df['rating'].value_counts().index)
plt.title('Rating vs. Type')
plt.xticks(rotation=90)
plt.show()
```
